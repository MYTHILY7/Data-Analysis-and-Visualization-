```python
import pytesseract
from pytesseract import Output
```

```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image_path = "/content/drive/MyDrive/MLDL05 - TECH SPARTANS/DAY 3/valid/00071c51c8e92a68_jpg.rf.7edd9de3e7a8e2431c6b54afd3796bd7.jpg"
image = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# License plate detection using Haar cascades
plate_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_russian_plate_number.xml")
plates = plate_cascade.detectMultiScale(gray, 1.1, 10)

for (x, y, w, h) in plates:
    roi = gray[y:y+h, x:x+w]  # Define the region of interest (roi)
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

    # Now you can use pytesseract on the roi
    text = pytesseract.image_to_string(roi, config='--psm 8')
    print("Detected license plate:", text)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
